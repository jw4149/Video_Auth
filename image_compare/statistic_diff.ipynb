```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
frame_id = 20
with open(f'diff_vecs/diff_{str(frame_id)}.txt', "rb") as f:
    diff_vectors = f.readlines()
diff = []
for i in range(len(diff_vectors)):
    diff_vectors[i] = diff_vectors[i].decode('utf-8')
    diff_vectors[i] = diff_vectors[i][1:-2].split(',')
    for j in range(len(diff_vectors[i])):
        diff_vectors[i][j] = float(diff_vectors[i][j]) 
```

```python
diff_vectors = np.array(diff_vectors)
diff_sum = np.sum(diff_vectors, axis=1)
print(diff_sum)
```

```python
plt.plot(diff_sum)
```

```python
print(diff_sum.mean(), diff_sum.var())
```

```python
diff_count = np.zeros(100)
bigger = 0
for x in diff_sum:
    if x > 0:
        diff_count[int(x)] += 1
    if x > 50:
        bigger += 1
print(bigger)
plt.plot(diff_count)
plt.xlabel("diff_sum")
plt.ylabel("count")
print(diff_count)
```
